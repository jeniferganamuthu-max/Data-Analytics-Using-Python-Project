```python
Sales=[10000,12000,25000,62000,98000]

total=sum(Sales)
avg=total/len(Sales)
high_Sales=max(Sales)
low_Sales=min(Sales)

print("Total:",total)
print("Average:",avg)
print("high_Sales:",high_Sales)
print("low_sales:",low_Sales)
```

```python
scores=[85,98,62,56,73]

total=sum(scores)
average=total/len(scores)
highest=max(scores)
lowest=min(scores)

print("Total Score:",total)
print("Average Score:",average)
print("Highest Score:",highest)
print("Lowest Score:",lowest)
```

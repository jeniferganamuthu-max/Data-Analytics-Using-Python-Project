## Data Filtering

```python
import pandas as pd

data = {
    "Region":["South","North","East","West","South","North"],
    "Product":["Laptop","Mobile","Laptop","Tablet","Mobile","Tablet"],
    "Sales":[50000,30000,45000,25000,20000,35000]
}

df = pd.DataFrame(data)

print(df)
```

```python
south_sales = df[df["Region"] == "South"]

print(south_sales)
```

```python
high_sales = df[df["Sales"] > 30000]

print(high_sales)
```

```python
sorted_df = df.sort_values("Sales")

print(sorted_df)
```

```python
region_sales = df.groupby("Region")["Sales"].sum()

print(region_sales)
```

```python
product_sales = df.groupby("Product")["Sales"].sum()

print(product_sales)
```

```python
print(df["Sales"].max())
```

```python
print(df["Sales"].min())
```

```python
print(df["Sales"].count())
```

```python
result = df.groupby("Region")["Sales"].agg(
    ["sum","mean","max","min","count"]
)
print(result)
```

Practice:Region-Wise Sales Report

```python
import pandas as pd

data = {
    "Region":["North","South","East","West","North","South","East","West","North","South"],
    "Product":["Mobile","Laptop","Tablet","Mobile","Laptop","Tablet","Mobile","Laptop","Mobile","Tablet"]
    ,
    "Sales":[10000,15000,22000,98000,50000,54000,68000,85000,76000,88000],
    "Quantity":[5,10,4,6,9,7,5,11,8,5]
}

df = pd.DataFrame(data)

east_sales = df[df["Region"] == "East"]
print(east_sales)

sorted_df = df.sort_values("Sales")
print(sorted_df)

region_sales = df.groupby("Region")["Sales"].sum()
print(region_sales)

result = df.groupby("Region")["Sales"].agg(
    ["sum","mean","max","min","count"]
)
print(result)
```

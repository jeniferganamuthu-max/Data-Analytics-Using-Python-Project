```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May"]

sales = [10000,12000,15000,13000,18000]

plt.plot(months,sales)
plt.title("Monthly Sales")

plt.xlabel("Months")

plt.ylabel("Sales Amount")

plt.show()
```

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May"]

sales = [10000,12000,15000,13000,18000]

plt.plot(months,sales,
         color ="green",
         marker="o",
         linestyle="--")
plt.title("Sales Trend")

plt.xlabel("Month")

plt.ylabel("Sales")

plt.show()
```

```python
import matplotlib.pyplot as plt

students =["A","B","C","D"]

marks = [80,90,75,95]

plt.bar(students,marks,
        color = ["red","blue","green","yellow"])

plt.title("Student Marks")

plt.xlabel("Students")

plt.ylabel("Marks")

plt.show()
```

```python
import matplotlib.pyplot as plt

months = ["Jan","Feb","Mar","Apr","May","Jun"]

sales = [10000,12000,15000,13000,18000,20000]

plt.figure(figsize=(18,5))

#Line Chart
plt.subplot(1,3,1)

plt.plot(months,
         sales,
         marker="o")

plt.title("Sales Trend")

plt.xlabel("Months")

plt.ylabel("Sales")

#Bar Chart
plt.subplot(1,3,2)

plt.bar(months,
        sales,
        color="blue")

plt.title("Sales Comparison")

plt.xlabel("Months")

plt.ylabel("Sales")

#pie chart
plt.subplot(1,3,3) #charged to 2 row, 3 column, position 3
expenses = [5000,3000,2000]

categories = ["Food","Travel","Shopping"]

plt.pie(expenses,
        labels=categories , autopct="%1.1f%%")

plt.title("Expense Distribution")

plt.tight_layout()
plt.show()
```
